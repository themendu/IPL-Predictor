```python
def tanh_scaler(data):
    data=data.drop(columns=['winner'],inplace=False)
    for c in data.columns:
        unnormalizedData = np.array(data[str(c)])
        m = np.mean(unnormalizedData, axis=0) 
        std = np.std(unnormalizedData, axis=0) 
        data[str(c)] = 0.5 * (np.tanh(0.01 * ((unnormalizedData - m) / std)) + 1)
    return data
```
